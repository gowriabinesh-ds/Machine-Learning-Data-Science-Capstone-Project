# Customer Churn Prediction - Capstone Project

### This dataset contains customer data from a telecom company and is used to analyze why customers leave the service (churn).

It includes information about:
- Customer details (gender, senior citizen, etc.)
- Services used (internet, phone, streaming)
- Account information (tenure, contract type, payment method)
- Billing details (monthly charges, total charges)

The dataset has **7,043 records**, and the target variable is **`Churn`**:
- `Yes` → Customer left the service  
- `No` → Customer stayed  

### Goal of this Analysis

- Understand customer behavior  
- Identify factors affecting churn  
- Build a model to predict which customers are likely to leave  

### Why this is important

By understanding churn, businesses can:
- Reduce customer loss  
- Improve retention strategies  
- Make better business decisions

## 3. Cleaning the Dataset

#### a. Removing the redundunt rows of data

```python
# Dropping columns which are not required for further analysis
dataset.drop(['customerID'], axis = 1, inplace=True)
```

#### b. Fixing data types

```python
# Convert TotalCharges to numerical
dataset['TotalCharges'] = pd.to_numeric(dataset['TotalCharges'], errors='coerce')

# Convert SeniorCitizen to categorical
dataset['SeniorCitizen'] = dataset['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
```

#### c. Handling missing values

```python
# Checking for Null records
dataset.isnull().sum()
```

```python
# Fill TotalCharges missing
# TotalCharges can have very high values (long-term customers) which are outliers
# So, using median because it is not affected by extreme values (outliers), it gives a more reliable central value than mean.

dataset['TotalCharges'].fillna(dataset['TotalCharges'].median(), inplace=True)
```

```python
dataset.isnull().sum()
```

#### d. Checking unique values in each column

```python
for col in dataset.columns:
    print("Column:", col)
    print(dataset[col].unique())
    print("----------------------")
```

#### e. Fixing inconsistent categories

```python
cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in cols:
    dataset[col] = dataset[col].replace({'No internet service': 'No'})

dataset['MultipleLines'] = dataset['MultipleLines'].replace({'No phone service': 'No'})
```

#### f. Creating a useful feature

```python
# To capture customer spending behavior over time, creating a new feature ‘AvgMonthlySpend’
# This calculates the average amount a customer spends per month based on their total charges and duration with the company
# Some customers have tenure = 0, to avoid division by zero errors, adding 1 to tenure

dataset['AvgMonthlySpend'] = dataset['TotalCharges'] / (dataset['tenure'] + 1)

dataset.head()
```

```python
# Round to 2 decimal places
dataset['AvgMonthlySpend'] = round(dataset['AvgMonthlySpend'], 2)

dataset.head()
```
